#### 1) Carregue a base e garanta que a base está como deveria.

Considere a base que você ajustou na lição passada. Carregue-a. Caso ainda haja alguma pendência, ajuste - lembre-se de que o scikitlearn não recebe variáveis em formato string, somente numéricas, e não aceita '*missings*'. 

- Separe 70% da base para treinamento e 30% para validação. Cada uma dessas partes terá dois objetos, um armazenando a variável resposta ```mau```e outro armazenando as variáveis explicativas (lembrando: sem variáveis string, já com as dummies).

````python
# Importando a biblioteca Pandas
import pandas as pd
````

````python
# Lendo a base de dados
df = pd.read_csv("new_demo01.csv")
````

````python
# Mostrando as primeiras 5 linhas da base
df.head()
````

````python
#Verificando se a base precisa de mais ajustes
df.info()
````

````python
# Neste bloco vamos separar a base para o treino e o teste

# Importando a função Train_test_split
from sklearn.model_selection import train_test_split

# Separando a a base (df) entre features e label
x = df.drop(columns = ['mau'])
y = df['mau']

# Separando a base de teste e a treinamento
x_train, x_test, y_train, y_test = train_test_split(
    x, 
    y, 
    test_size = 0.3,     # separa 30% para teste
    stratify = y,        # Mantém a proporção da classe 'mau'
    random_state = 42    # reprodutibilidade
)
````

````python
# Verificando a divisão
print(f"Treino: {x_train.shape[0]} amostras")
print(f"Teste: {x_test.shape[0]} amostras")
print("\nProporção da classe 'mau' no treino:", y_train.mean())
print("Proporção da classe 'mau' no teste:", y_test.mean())
````

#### 2) Vamos para o modelo:

1. Defina um objeto com a função da árvore de decisão vista em aula.
2. Treine o modelo com os dados que você separou para treinamento.
3. Visualize a árvore. Talvez você queira aumentar um pouco a figura.
4. Produza uma visualização da matriz de classificação (ou matriz de confusão) - coloque os rótulos como "aprovados" e "reprovados" - pois afinal de contas, são essas as decisões que se podem tomar com propostas de crédito.
5. Calcule a acurácia na base de treinamento

````python
# Definindo o modelo de Árvore de Decisão

# Importando a função DecisionTreeClassifier da biblioteca scikit-learn para criar o modelo
from sklearn.tree import DecisionTreeClassifier 

# Definir o modelo de árvore de decisão
modelo_arvore = DecisionTreeClassifier(random_state = 42)
````

````python
# Treinando o modelo
modelo_arvore.fit(x_train, y_train)
````

````python
# Construindo a arvore de decisão

# Importanto as bibliotecas
from sklearn.tree import plot_tree # importanto a função plot_tree da sklearn.tree
import matplotlib.pyplot as plt    # importanto a biblioteca e nomeando-a como plt

# Definindo o tamnho da figura
plt.figure(figsize = (20, 10))

# Plotar a árvore
plot_tree(
    modelo_arvore,
    filled = True,                            #Preenche os nós com cores
    feature_names = x.columns,                # Nomes das Features
    class_names = ['Aprovados','Reprovados'], # Nomes das classes
    rounded = True                            # Bordas arredondadas 
)

# Mostrar a árvore
plt.show()
````

````python
# Construindo a matrix de Confusão

# Importando funções da sklearn.metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Fazer previsões no conjunto de treino
y_pred_train = modelo_arvore.predict(x_train)

# Criar a matriz de confusão
matriz_confusao = confusion_matrix(y_train, y_pred_train)

# Exibir a matrix de confusão com rótulos
disp = ConfusionMatrixDisplay(
    confusion_matrix = matriz_confusao,
    display_labels = ['Aprovado','Reprovado']
)

# Definindo o estilo da matriz
disp.plot(cmap = plt.cm.Blues)                    # Define o estilo da matriz
plt.title ('Matriz de Confusão - Base de Treino') # Personliza o nome da figura

# Vizualizar a figura
plt.show()
````

````python
# Calculando a acurácia

# Importando a função
from sklearn.metrics import accuracy_score

# calcular a acurácia
acuracia = accuracy_score(y_train, y_pred_train)
print(f'Acurácia na base de treino: {acuracia: .2%}')
````

#### 3) Vamos avaliar o modelo na base de testes

1. Classifique a base de teste de acordo com a árvore que você treinou no item 2.
2. Produza a visualização da matriz de confusão para a base de teste.
3. Calcule a acurácia da base de teste. Compare com a acurácia da base de treinamento.

5. Como ficaria a acurácia se você classificasse todos os contratos como 'bons'?

````python
# Faça previsões na base de teste
y_pred_test = modelo_arvore.predict(x_test)
````

````python
# Criando a matriz de confusão

# Crie a matrix de confusão
matriz_confusao = confusion_matrix(y_test, y_pred_test)

# Exiba a matriz de confusão com rótulos
disp = ConfusionMatrixDisplay(
    confusion_matrix = matriz_confusao,
    display_labels = ['Aprovados','Reprovados']
)

# Definindo o estilo da matriz
disp.plot(cmap = plt.cm.Blues)                  # Defini o estilo da matriz
plt.title('Matriz de Confusão - Base de Teste') # Personaliza o nome da figura

# Exiba a matriz e a salve a figura no diretório
plt.savefig('Matriz - Confusao - Teste')
plt.show()
````

````python
# Calcular a acurácia na base de teste
acuracia_teste = accuracy_score(y_test, y_pred_test)
print(f"Acurácia na base de teste: {acuracia_teste: .2%}")

# Compare com a acurácia na base de treino
acuracia_treino = accuracy_score(y_train, y_pred_train)
print(f'Acurácia na base de treino: {acuracia_treino: .2%}')
````

````python
# Treine uma nova ávore com os hiperparâmetros ajustados

# Defina uma nova árvore com hiperparâmetros ajustados
nova_arvore = DecisionTreeClassifier(
    min_samples_leaf = 5, # Defini mínimo de observações por folha
    max_depth = 10,       # Defini a profundidade máxima da árvore
    random_state = 123    # Para garantir reprodutibilidade
)

# Treine uma nova árvore
nova_arvore.fit(x_train, y_train)

# Faça previsões na base de teste
y_pred_nova_arvore = nova_arvore.predict(x_test)
````

````python
# Definindo matriz de confusão para a árvore nova

# Defina uma matriz de confusão
matriz_confusao_nova = confusion_matrix(y_test, y_pred_nova_arvore)

# Exiba a matriz de confusão com rótulos
disp = ConfusionMatrixDisplay(
    confusion_matrix = matriz_confusao_nova,
    display_labels = ['Aprovado','Reprovado']
)

# Definindo o estilo da figura
disp.plot(cmap = plt.cm.Blues)                             # Defini o estilo de figura
plt.title('Matriz de Confusão - Nova Árvore (Base Teste)') # Defini o título da figura

# Mostre a figura e salve no diretório
plt.savefig('Matriz - Confusao - Teste - Nova')
plt.show()
````

````python
# Calculando a Proporção de proponentes clasificados como 'maus'

# Calcule a proporção de proponentes
proporcao_maus = sum(y_pred_nova_arvore) / len(y_pred_nova_arvore)

# Mostre a proporção
print(f'Proporção de proponentes Classificados como "maus": {proporcao_maus: .2%}')
````

````python
# Calcular a acurácia na base de TESTE para o novo modelo
acuracia_teste_nova = accuracy_score(y_test, y_pred_nova_arvore)
print(f"Acurácia na base de teste (nova árvore): {acuracia_teste_nova:.2%}")
````

````python
# Calculando a acurácia se todos os contratos fossem calculados como bons

# Classifiqe todos como 'bons'
y_pred_todos_bons = [0] * len(y_test)

# Calcule a acurácia
acuracia_todos_bons = accuracy_score(y_test, y_pred_todos_bons)

# Mostre acurácia
print(f'Acurácia se todos fossem classificados como "bons" {acuracia_todos_bons: .2%}')
````
