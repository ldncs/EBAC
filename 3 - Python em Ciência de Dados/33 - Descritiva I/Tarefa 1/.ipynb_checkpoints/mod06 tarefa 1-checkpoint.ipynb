```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Módulo 6 tarefa 1

#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
df = pd.read_csv('sinasc_RO_2019.csv', sep = ',', usecols = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS'])
```

```python
#mostre a quantidade de linhas e colunas
df.shape
```

```python
# mostre uma visão mais detalhada do df
df.info()
```

```python
# mostre o Data Frame
df.head()
```

```python
#Identifique linhas que tenham pelo menos um Nan
dados_nan = df.isna().any(axis=1)

#Filtre linhas que tenha pelo menos um Nan
df_no_nan = df[~dados_nan]
```

```python
#Mostre como ficou o novo df
df_no_nan.info()
```

```python
#mostre o novo data frame
df_no_nan.head(10)
```

```python
# monte um dicionário do estado civil da gestante
dic_estcivmae = {1: 'Solteira', 
                 2: 'Casada', 
                 3: 'Viúva', 
                 4: 'Separada judicialmente / Divorciada', 
                 5: 'União consensual',
                 9: 'Ignorado'
}

#subtitua os valores da coluna ESTCIVMAE
df_no_nan.loc[:, 'ESTCIVMAE'] = df_no_nan['ESTCIVMAE'].replace(dic_estcivmae)

# mostre os primeiros 10 linhas da coluna ESTCIVMAE
df_no_nan['ESTCIVMAE'].head(10)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Criar o gráfico para cada uma das variáveis

# Configurar o tamanho da figura
plt.figure(figsize=(14, 20))

# Subplot 1: APGAR1 por SEXO
plt.subplot(3, 2, 1)
sns.barplot(x='SEXO', y='APGAR1', data=df_no_nan, hue='SEXO', palette='Set2', legend=False)
plt.title('APGAR 1 por SEXO')

# Subplot 2: APGAR5 por SEXO
plt.subplot(3, 2, 2)
sns.barplot(x='SEXO', y='APGAR5', data=df_no_nan, hue='SEXO', palette='Set2', legend=False)
plt.title('APGAR 5 por SEXO')

# Subplot 3: APGAR1 por GESTACAO
plt.subplot(3, 2, 3)
sns.barplot(x='GESTACAO', y='APGAR1', data=df_no_nan, hue='GESTACAO', palette='Set2', legend=False)
plt.title('APGAR 1 por GESTACAO')
plt.xticks(rotation=90)

# Subplot 4: APGAR5 por GESTACAO
plt.subplot(3, 2, 4)
sns.barplot(x='GESTACAO', y='APGAR5', data=df_no_nan, hue='GESTACAO', palette='Set2', legend=False)
plt.title('APGAR 5 por ESTCIVMAE')
plt.xticks(rotation=90)

# Subplot 5: APGAR1 por ESTCIVMAE
plt.subplot(3, 2, 5)
sns.barplot(x='ESTCIVMAE', y='APGAR1', data=df_no_nan, hue='ESTCIVMAE', palette='Set2', legend=False)
plt.title('APGAR 5 por ESTCIVMAE')
plt.xticks(rotation=90)

# Subplot 6: APGAR5 por ESTCIVMAE
plt.subplot(3, 2, 6)
sns.barplot(x='ESTCIVMAE', y='APGAR5', data=df_no_nan, hue='ESTCIVMAE', palette='Set2', legend=False)
plt.title('APGAR 5 por ESTCIVMAE')
plt.xticks(rotation=90)

# Ajustar layout para visualização
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Calcular a média do APGAR5 por CONSULTAS
mean_apgar5 = df_no_nan.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(8,6))
sns.barplot(x='CONSULTAS', y='APGAR5', hue = 'CONSULTAS', data=mean_apgar5, palette='viridis', legend = False)

# Adicionar títulos e rótulos
plt.title('Média do APGAR 5 por CONSULTAS', fontsize=14)
plt.xlabel('Consultas', fontsize=12)
plt.ylabel('Média APGAR 5', fontsize=12)

# Exibir o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Configurar o tamanho da figura
plt.figure(figsize=(14, 10))

# Subplot 1: Frequência de SEXO
plt.subplot(2, 2, 1)
sns.countplot(x='SEXO', data= df_no_nan, hue = 'SEXO', palette='Set2', legend = False)
plt.title('Frequência de SEXO')

# Subplot 2: Frequência de CONSULTAS
plt.subplot(2, 2, 2)
sns.countplot(x='CONSULTAS', data= df_no_nan, hue = 'CONSULTAS', palette='Set2', legend = False)
plt.title('Frequência de CONSULTAS')

# Subplot 3: Frequência de ESTCIVMAE
plt.subplot(2, 2, 3)
sns.countplot(x='ESTCIVMAE', data= df_no_nan, hue = 'ESTCIVMAE', palette='Set2', legend = False)
plt.title('Frequência de ESTCIVMAE')
plt.xticks(rotation=90)

# Subplot 4: Frequência de GESTACAO
plt.subplot(2, 2, 4)
sns.countplot(x='GESTACAO', data= df_no_nan, hue = 'GESTACAO', palette='Set2', legend = False)
plt.title('Frequência de GESTACAO')
plt.xticks(rotation=90)

# Ajustar layout para visualização
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
