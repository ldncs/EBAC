```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sinasc_RO_2019.csv', sep = ',', usecols = ['PESO', 'IDADEMAE'])
```

```python
#mostre a quantidade de linhas e colunas
df.shape
```

```python
# mostre uma visão mais detalhada do df
df.info()
```

```python
# mostre o Data Frame
df.head(10)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# criar boxplot idade da mãe 
plt.figure(figsize = (10, 6))
sns.boxplot(x=df['IDADEMAE'], color = 'skyblue')

# personalizar o gráfico
plt.title('Boxplot da idade da mãe', fontsize = 16)
plt.xlabel('Idade da mãe', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)

# exibir gráfico
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# criar o boxplot do peso do bebê
plt.figure(figsize = (10, 6))
sns.boxplot(x = df['PESO'], color = 'skyblue')

# personalizar o gráfico
plt.title('Boxplot do Peso do Bebê', fontsize = 16)
plt.xlabel('Peso do bebê', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)

# exibir o gráfico
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Criar o histograma da idade da mãe
plt.figure(figsize=(10, 6))
sns.histplot(df['IDADEMAE'], bins=20, kde = False, color='skyblue', edgecolor='black')

# Personalizar o gráfico
plt.title('Histograma da Idade da Mãe', fontsize=14)
plt.xlabel('Idade da Mãe', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# exibir o gráfico
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# criar histograma peso do bebê
plt.figure(figsize = (10, 6))
sns.histplot(df['PESO'], bins = 20, kde = False, color = 'skyblue', edgecolor = 'black')

# Personalizar o gráfico
plt.title('Histograma Peso do Bebê', fontsize = 16)
plt.xlabel('Peso do Bebê', fontsize = 12)
plt.ylabel('Frequência', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)

# Exebir o Gráfico
plt.show()
```
