#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# construa uma função para categorizar o apgar5
def categorize_apgar5(value):
    if 0 <= value <= 3:
        return 'asfixia grave'
    elif 4 <= value <= 5:
        return 'asfixia moderada'
    elif 6 <= value <= 7:
        return 'asfixia leve'
    elif 8 <= value <= 10:
        return 'normal'
    else:
        return 'Valor fora do intervalo'


# aplicando função a coluna apgar5 do df para as categoria
df_no_nan['APGAR5'] = df_no_nan['APGAR5'].apply(categorize_apgar5)

# mostre a nova coluna apgar5 do df
df_no_nan['APGAR5']
```

```python
# verificando os dados unicos da coluna apgar5 do df
df_no_nan['APGAR5'].unique()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# filtre os dados para os SEXO
masculino = df_no_nan[df_no_nan['SEXO'] == 'Masculino']
feminino = df_no_nan[df_no_nan['SEXO'] == 'Feminino'] 

# criando a figura
plt.figure(figsize = (10, 15))

# personalize gráfico
plt.subplot(2, 1, 1)
sns.histplot(x = 'GESTACAO', y = 'APGAR5', hue = 'GESTACAO', data = masculino, palette = 'flare', legend = False)
plt.title('Distribuição APGAR5 masculino por tempo de Gestação', fontsize = 16)
plt.xlabel('Gestação', fontsize = 12)
plt.ylabel('APGAR5', fontsize = 12)
plt.xticks(rotation = 90)

# personalize gráfico
plt.subplot(2, 1, 2)
sns.histplot(x = 'GESTACAO', y = 'APGAR5', hue = 'GESTACAO', data = feminino, palette = 'flare', legend = False)
plt.title('Distribuição APGAR5 feminino por tempo de Gestação', fontsize = 16)
plt.xlabel('Gestação', fontsize = 12)
plt.ylabel('APGAR5', fontsize = 12)
plt.xticks(rotation = 90)

# ajuste o espaço entre os gráficos
plt.subplots_adjust(wspace = 1.00, hspace = 0.5)

# exiba o gráfico
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# faça uma tabela cruzada entre apgar5 e gestacao
tabela_cruzada = pd.crosstab(df_no_nan['GESTACAO'], df_no_nan['APGAR5'])

# exiba a tabela cruzada
tabela_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Crie uma tabela com a porcentagem dos valores cruzados gestacao e apgar5
procent_tabela_cruzada = tabela_cruzada.div(tabela_cruzada.sum(axis = 1), axis = 0)*100
(procent_tabela_cruzada).round(2)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Filtrar dados de partos prematuros (gestação <= 37 semanas)
prematuros = df_no_nan[df_no_nan['GESTACAO'].isin(['32 a 36 semanas', '28 a 31 semanas', '22 a 27 semanas', 'Menos de 22 semanas'])]

# Filtrar dados de partos não prematuros (gestação > 37 semanas)
n_prematuros = df_no_nan[df_no_nan['GESTACAO'].isin(['37 a 41 semanas', '42 semanas e mais'])]
```

```python
# Criação da figura e subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 15))

# Personalizando o primeiro gráfico
sns.barplot(ax=axes[0], x='GESTACAO', y='APGAR5', hue='GESTACAO', data=prematuros, palette='flare', legend=False)
axes[0].set_title('APGAR5 por tempo de Gestação de Prematuros', fontsize=16)
axes[0].set_xlabel('', fontsize=12)
axes[0].tick_params(axis='x', rotation=90)

# Personalizando o segundo gráfico
sns.barplot(ax=axes[1], x='GESTACAO', y='APGAR5', hue='GESTACAO', data=n_prematuros, palette='flare', legend=False)
axes[1].set_title('APGAR5 por tempo de Gestação Não Prematuros', fontsize=16)
axes[1].set_xlabel('', fontsize=12)
axes[1].tick_params(axis='x', rotation=90)

# Ajustando o espaço entre os gráficos
plt.subplots_adjust(wspace=1.00, hspace=0.5)

# Adicionando rótulos gerais para os eixos
fig.supxlabel('Tempo de Gestação', fontsize=14)
fig.supylabel('Pontuação APGAR5', fontsize=14)

# Exibindo o gráfico
plt.show()
```

É possível perceber que no gráfico dos bebês prematuros existem uma maior incidência de anormalidade no estado clínico dos bebês nos primeiros minutos de vida. Ainda, há de se observar que existe uma maior variabilidade nos bebês prematuros igual ou menor que 22 semanas de gestação.

Já no gráfico dos bebês que não são considerados prematuros existe uma maior incidência de normalidade entre os bebês. No entatno, para os bebês que ultrapassam as 41 semanas de gestação existe uma maior variabilidade nos dados, o que pode indicar que ultrapassar o período de 31 a 41 semanas pode influenciar na normalidade clínica dos bebês. 

Portanto, os gráficos corroboram com a afirmativa dos médicos, pois os dados mostram que quanto menor é o tempo de gestação, maior a chance de haver anormalidades nos primeiros minutos de vida do bebê.

```python
# Correlação entre APGAR5 e Peso do Bebê
correlacao = df['PESO'].corr(df['APGAR5'])
print(f'Correlação entre APGAR5 e Peso do Bebê: {correlacao:.2f}')
```

Uma correlação de 0.17 significa que embora haja uma relação positiva entre o peso do bebê e APGAR5 essa relação é muito fraca. Portanto, seria necessário averiguar outras variáveis. 

```python
# Crie faixas de peso
bins = [0, 2500, 4000, 5000, 7000]
labels = ['Muito Baixo Peso', 'Peso Normal', 'Acima do Peso', 'Peso Muito Elevado']
df_no_nan.loc[:, 'FAIXA_PESO'] = pd.cut(df_no_nan['PESO'], bins=bins, labels=labels)

# Crie a figura
fig, axes = plt.subplots(2, 1, figsize=(10, 15))

# Personalizando o primeiro gráfico (Scatterplot)
sns.scatterplot(x='APGAR5', y='PESO', data=df_no_nan, hue = 'FAIXA_PESO', alpha=0.7, ax=axes[0], palette = 'flare')
axes[0].set_title('Relação entre Peso do Bebê e APGAR5', fontsize=16)
axes[0].set_xlabel('Peso do Bebê (gramas)', fontsize=12)
axes[0].set_ylabel('APGAR5', fontsize=12)

# Personalizando o segundo gráfico (Boxplot)
sns.boxplot(x='PESO', y='APGAR5', hue = 'FAIXA_PESO', data=df_no_nan, palette='flare', ax=axes[1])
axes[1].set_title('APGAR5 por Faixa de Peso do Bebê', fontsize=16)
axes[1].set_xlabel('Peso do Bebê (Em gramas)', fontsize=12)
axes[1].set_ylabel('APGAR5', fontsize=12)
axes[1].tick_params(axis='x', rotation=90)  # Use tick_params para rotacionar

# Ajustando o espaço entre os gráficos
plt.subplots_adjust(wspace=0.5, hspace=0.5)

# Salvar a figura
#plt.savefig('figura_peso_apgar5.png', dpi=300, bbox_inches='tight')  # Salva em alta resolução

# Exibir gráficos
plt.show()
```

Também não é possível encontrar uma relação visual entre as duas variáveis por meio dos gráficos. Portanto, **não pode-se afirmar** que a relação entre o **peso do bebê interefere em seu estado clínico de saúde nos primeiros minutos de vida.**
